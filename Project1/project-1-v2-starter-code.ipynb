## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
### 7. Calculate player "power"

def cal_power(players_, pokedex_, play_ID):
    """ takes in 3 inputs and calculates a player's power
    
        Inputs: 
        
        players_, as a dictionary with all players' information
        pokedex_ ,as a dictionary with all pokemon's information
        play_ID as, an int, ID of the player to find the power for
        
        Function:
        1) Runs through all the pokemons a player has
        2) For each pokemon sum up all the stats (e.g. atk, def, sa, sd, speed, hp)
        3) Prints player's power and returns power information
    """
    power = 0
    for pokemon in players_[play_ID]['player_pokemon']:
        power += pokedex[pokemon]['hp']
        power += pokedex[pokemon]['attack']
        power += pokedex[pokemon]['defence']
        power += pokedex[pokemon]['specialattack']
        power += pokedex[pokemon]['specialdefence']
        power += pokedex[pokemon]['speed']
        
    print("[PlayerID={ID}] {player}'s power is {power}.".format(player=players_[play_ID]['player_name'], ID=play_ID, power=str(power)) )
    return power
    
cal_power(players, pokedex, 1)
cal_power(players, pokedex, 2)
````

````python
### 8.1 Load a pokedex file containing all the pokemon
def cleanfile(lis):
    """
        Takes in a lis and outputs another metrics list based on the following criteria:
    
        First For loop: 
            1) Quotes are removed from each cell item
            2) Each line as then converted into its own list with elements seperated by ','
            3) lines are added as list of a list to the <cleaner> variable
        
        Second For loop: 
            1) Reimagined as a 2-Diamension matrix
            2) For each element in each list of the general list
                2.1) Check if element is blank string, convert them to -1 
                2.2) Else Try to change element data type to float
                    2.2.1) If can't, skip, leave it as a sting
        
        Prints & Returns Output [cleaner] as a list of list
    """
    cleaner = []
    
    for line in lis:
        line = line.replace('"','')
        line = line.split(',')
        cleaner += [line]
        
    for rows in range(len(cleaner)):
        for cols in range(len(cleaner[rows])):
            
            if cleaner[rows][cols] == '':
                cleaner[rows][cols] = -1
            else:
                try: 
                    cleaner[rows][cols] = float(cleaner[rows][cols])
                except:
                    pass

    print(cleaner)
    return cleaner
cleanfile(clean1)
````

````python

def pokedex_(lis):
    """
        Using the cleaned pokedex list create a dictionary for pokemon with the following key pairs:
        pokedexnumber : unique identifier for each pokemon (integer)
        name : the name of the pokemon (string)
        type : the category of pokemon (string)
        hp : base hitpoints (integer)
        attack : base attack (integer)
        defense : base defense (integer)
        specialattack : base special attack (integer)
        specialdefense : base sepecial defense (integer)
        speed : base speed (integer)
    
        Runs 2 loops: 
            1) One creates a distionary of key value pairs for each stat for one pokemon
            2) Second out loop stitches the pokemons in to a distionary of pokemon:stats key pairs
    
    
    """
    dict1 = {}
    for pokemon in lis[1:]:
        dict2 = {}
        for i,stats in enumerate(pokemon):
            dict2[lis[0][i].lower()] = stats
        dict1[pokemon[0]] = dict2
    

    return dict1

pokedex_(clean2)
        
````

````python
import random

def poke_filter(pokedex_data, filter_options):
    """
        Take the parsed pokedex information as an argument.
        Take a dictionary as a parameter with keys matching the features of the Pokedex, filtering by exact match for string type values, and/or filter continuous variables specified value that is greater than or equal to the dictionary key parameter.
        Return multiple elements from the Pokedex

        Input Values:
            1) filter_options as a dictionary, WORDS FOR <<<KEYS>>> IN LOWERCAPS
                1.1) <<<CAPS>>> for first letter for <<<VALUES>>> of Keys
                    1.1.1) Example: 
                                    filter_options = {
                                    'attack':   25,
                                    'defense':  30,
                                    'type': 'Electric',
                                    'name': 'Pikachu'
                                    }
        
        
        In Function Variebles: 
            1) pokeout, as dictionary - stores the filtered pokemon information after the the filter is applied to the pokedex
            2) dictkey, as key (<<<--- whatever the datatype the key turns put to be)
            
        Function walkthrough:
            1) A recursive function was written. Task done through use of recursion
            2) If the filter is a blank dictionary, return the pokedex_data input
            3) Randomly pulls out a key <dictkey> from the filter <filter_options>
            4) Filters the pokedex using the value from the key
                4.1) If the value is a string, find exact match - accept pokemon and saves to pokeout if accepted
                4.2) Else accept if equal or greater than -  accept pokemon and saves to pokeout if accepted
            5) Delete the key from the dictionary
            6) Have the function call itself again with pokeout, filter_options (<<<--- popped) as inputs
            
    """
    if len(filter_options) > 0:
        
        pokeout = {}
        dictkey = random.choice(list(filter_options))
        
        
        for pokeid in pokedex_data:
            
            if isinstance(filter_options[dictkey], str):
                if pokedex_data[pokeid][dictkey] == filter_options[dictkey]:
                    pokeout[pokeid] = pokedex_data[pokeid]
                    
            elif pokedex_data[pokeid][dictkey] >= filter_options[dictkey]:
                pokeout[pokeid] = pokedex_data[pokeid]
                
        filter_options.pop(dictkey)
        return poke_filter(pokeout, filter_options)
    
    
    else:
        lis = [] 
        for pokemon in pokedex_data:
            lis += [pokedex_data[pokemon]]
        return lis
    

pokedex_data = pokedex_(clean2)
filter_options = {
    'attack':   25,
    'defense':  30,
    'type': 'Electric',
    'name': 'Pikachu'
}

newdict = poke_filter(pokedex_data, filter_options)
print(newdict)

    
            
        
````
